```python

KEY = {}
NAVER_AUTH = 'NAVER_LAND_AUTH'
with open('key', 'r') as f:
    for line in f:
        if len(line.strip()) == 0:
            continue
        key, value = line.strip().split(':')
        KEY[key] = value
```

```python
import requests
import json
import pandas as pd
down_url = 'https://new.land.naver.com/api/complexes/8928'
rep = requests.get(down_url,data={"sameAddressGroup":"false"},headers={
    "Accept-Encoding": "gzip",
    "Host": "new.land.naver.com",
    "Referer": "https://new.land.naver.com/complexes/8928?ms=37.482968,127.0634,16&a=APT&b=A1&e=RETAIL",
    "Sec-Fetch-Dest": "empty",
    "Sec-Fetch-Mode": "cors",
    "Sec-Fetch-Site": "same-origin",
    "authorization": KEY[NAVER_AUTH],
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36"
})
print(f'rep : {rep}')
rep.encoding = "utf-8-sig"
data_js = json.loads(rep.text)

print(f'key : [{data_js.keys()}]')

print(data_js['complexDetail'])
print(data_js['complexPyeongDetailList'])

apt_data=data_js["complexDetail"]
pyoeng_list=apt_data["pyoengNames"].split(", ")
apt_data_pd=pd.DataFrame(data=apt_data,index=range(len(pyoeng_list)))
```
